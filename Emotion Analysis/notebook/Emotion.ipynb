# 📈 Emotion Distribution

A count plot is used to visualize the frequency of each emotion.

This helps identify:

- Most frequent emotions
- Least frequent emotions
- Overall class balance

```python

# Read CSV
df = pd.read_csv(r"C:\Users\MAHE\Downloads\emotion.csv")

# Count occurrences of each emotion
emotion_count = df['Emotion'].value_counts()

# Pie chart
plt.figure(figsize=(6,6))

plt.pie(
    emotion_count,
    labels=emotion_count.index,
    autopct='%1.1f%%'
)

plt.title("Emotion Distribution")

plt.show()

```

# ☁️ Word Cloud Visualization

A Word Cloud provides a visual representation of the most frequently occurring words within the dataset.

Larger words indicate higher frequency and help highlight dominant themes present in the text.

```python
# Combine all text from a column
text = " ".join(df['Comment'].astype(str))

# Generate word cloud 
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='lightblue'
).generate(text)

# Display
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
