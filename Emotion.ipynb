```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import seaborn as sns

# Load CSV
df = pd.read_csv(r"C:\Users\MAHE\Downloads\emotion.csv")
print(df.head())
print(df.info())
print(df.describe())
print(df.columns)
# Combine all text from a column
text = " ".join(df['Comment'].astype(str))

# Generate word cloud 
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='lightblue'
).generate(text)

# Display
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt

# Read CSV
df = pd.read_csv(r"C:\Users\MAHE\Downloads\emotion.csv")

# Count occurrences of each emotion
emotion_count = df['Emotion'].value_counts()

# Pie chart
plt.figure(figsize=(6,6))

plt.pie(
    emotion_count,
    labels=emotion_count.index,
    autopct='%1.1f%%'
)

plt.title("Emotion Distribution")

plt.show()

```
